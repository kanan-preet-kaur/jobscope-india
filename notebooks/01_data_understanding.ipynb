### Leading/Trailing Spaces

```python
space_counts = {
    col: (df[col].astype(str) != df[col].astype(str).str.strip()).sum()
    for col in df.select_dtypes(include=["object", "string"]).columns
}

pd.DataFrame(space_counts.items(), columns=["Column", "Leading/Trailing Spaces"])
```

### Empty Strings

```python
empty_counts = {
    col: (df[col].astype(str).str.strip() == "").sum()
    for col in df.select_dtypes(include=["object", "string"]).columns
}

pd.DataFrame(empty_counts.items(), columns=["Column", "Empty Strings"])
```

## Initial Data Audit Report

### 1. Dataset Overview
- The dataset contains **97,929 job postings** across **17 attributes**.
- It includes information related to job titles, companies, salaries, required experience, locations, skills, company ratings, reviews, and detailed job descriptions.
- The dataset appears comprehensive and is suitable for performing exploratory data analysis, visualization, dashboard development, SQL querying, and further feature engineering.

---

### 2. Data Structure
- The dataset contains a combination of numerical and categorical features.
- Numerical columns include salary ranges, experience ranges, ratings, and review counts.
- Categorical columns include job titles, company names, locations, uploaded dates, skills, and job descriptions.

---

### 3. Missing Values
- Missing values are present in certain columns, particularly **ReviewsCount** and **AggregateRating**.
- These missing values will be examined during the data cleaning phase to determine the most appropriate treatment strategy.

---

### 4. Duplicate Records
- Duplicate records were inspected.
- No duplicate rows were identified during the initial audit.
